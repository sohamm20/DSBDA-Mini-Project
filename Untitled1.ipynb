```python
df1= data.drop(["Unnamed: 0"] , axis=1)
```

```python
df2= df1.drop(["Type"], axis=1)
```

```python
df2
```

```python
#Importing Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X= df2.drop("0 Type", axis=1)
y= df2["0 Type"]
```

```python
np.random.seed(1)
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
import gradio as gr
from sklearn import datasets
```

```python
def iris(Air_Temperature, Process_temperature, Rotational_speed, Torque,Tool_wear, Target):
    prediction = model.predict([[Air_temperature [K], Process_temperature [K],Rotational_speed [rpm],Torque [Nm],Tool_wear [min], Target]])
    prediction = data.target_names[prediction]
    return prediction
    
    
    
#Input


input2 = gr.inputs.Number(label="Air temperature [K]")
input3 = gr.inputs.Number(label="Process temperature [K]")
input4 = gr.inputs.Number(label="Rotational speed [rpm]")
input5 = gr.inputs.Number(label="Torque [Nm]")
input6 = gr.inputs.Number(label="Tool wear [min]")
input7 = gr.inputs.Number(label="Target")

#output
output = gr.outputs.Textbox(label= "Machine Failure status")

#Interface
gui = gr.Interface(fn=iris,
                   inputs=[input2, input3, input4, input5, input6, input7],
                   outputs=output).launch()
```
